$f(x) = \frac{x^2-4}{x-2}$

$\lim_{x \rightarrow 3} = \frac{x^2-4}{x-2}$

Substituindo o valor de x por 3 na expressão, temos:

$\lim_{x \rightarrow 3} = \frac{3^2-4}{3-2}$

$\lim_{x \rightarrow 3} = \frac{9-4}{1}$

$\lim_{x \rightarrow 3} = \frac{5}{1}$

$\lim_{x \rightarrow 3} = {5}$

A biblioteca MATPLOTLIB está sendo utillizada para plotar graficos.

A NUMPY está sendo utilizada para criar o array de caracteres.

A biblioteca SYMPY está sendo usada para calcular limites.

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy


def f(x):
    return (x**2 - 4)/(x - 2)


x = sympy.symbols('x')


limite_esquerda = sympy.limit(f(x), x, 3, dir='-')

print(f"Limite pela esquerda: {limite_esquerda}")


x = np.linspace(-10, 10, 1000)


y = f(x)


plt.plot(x, y)


plt.plot(3, 5, 'ro')


plt.grid(True)


plt.xlabel('x')
plt.ylabel('f(x)')


plt.show()
```

```python
import sympy
from fractions import Fraction


def limite1(valor):
  x = sympy.symbols('x')
  
  f = sympy.sqrt(x-8) +3

  
  limite = sympy.limit(f, x, valor)

  
  print(limite)

def limite2(valor):
  x = sympy.symbols('x')
  f = (sympy.sqrt(x)-3)/(x-9)
  limite = sympy.limit(f,x,valor)
  print(limite)

def limite3(valor):
  x = sympy.symbols('x')
  f = (2*x**2 - 5*x + 3)/(2*x**2 - 3*x) 
  limite = sympy.limit(f, x, valor)
  print(limite)

def limite4(valor):

  x = sympy.symbols('x')

  
  f = sympy.Piecewise((x**2 + 3, x >= -3), (-2*x + 6, x < -3))

  
  limite = sympy.limit(f, x, valor)

  print(limite)




limite1(9)
limite2(9)
limite3(Fraction(3/2))
limite4(-3)
```
